dado que en lesiones victimas y homicidios victimas el sexo esta escrito diferente en ambos, converitremos a un solo tipo de dato estas columnas. 

* reemplazo en sexo, del df lesiones victimas, 'varon' por 'masculino' y 'mujer' por 'femenino'. Luego, en df homicidios victimas reemplazo, 'MASCULINO' por 'masculino' y 'FEMENINO' por 'femenino'

```python
df_homicidios_victimas['sexo'].replace({'MASCULINO': 'masculino', 'FEMENINO': 'femenino'}, inplace = True)
df_lesiones_victimas['sexo'].replace({'Varon': 'masculino','VARON': 'masculino', 'varon':'masculino',
                                    'Mujer': 'femenino','mujer':'femenino', 'MUJER':'femenino'}, inplace = True) 
```

realizo un merge de las hojas victimas y hechos del libro homicidios y tambien, por otro lado, lo mismo para el libro lesiones

```python
merged_homicidios = pd.merge(df_homicidios_hechos, df_homicidios_victimas, left_on='id_hecho', right_on='id_hecho', how='inner')
```

```python
merge_lesiones = pd.merge(df_lesiones_hechos, df_lesiones_victimas, left_on='id_hecho', right_on='id_hecho', how= 'inner')
```
